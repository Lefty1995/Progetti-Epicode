# Sinossi del Progetto Finale  
## Synthetic Data Sandbox - VAE Edition

## 1. Introduzione

Il progetto ha l’obiettivo di creare una piccola piattaforma in grado di generare dati sintetici partendo da un dataset tabellare reale.

L’idea principale è quella di prendere un dataset con informazioni strutturate, come dati medici o finanziari, analizzarlo, prepararlo e usarlo per addestrare un modello di tipo VAE, cioè un Variational Autoencoder.

Il modello dovrà imparare la struttura generale dei dati originali e generare un nuovo dataset sintetico. Questo dataset non deve essere una copia dei dati reali, ma deve mantenere caratteristiche statistiche simili, come distribuzioni, correlazioni e comportamento generale delle variabili.

Il progetto è pensato come una sandbox, cioè un ambiente dove un utente può caricare un dataset, generare dati sintetici e confrontarli con quelli originali.

## 2. Obiettivo del Progetto

L’obiettivo principale è sviluppare un sistema che permetta di:

- caricare un dataset tabellare;
- preparare i dati per l’addestramento;
- addestrare un modello VAE;
- generare un dataset sintetico;
- confrontare i dati reali con quelli sintetici;
- verificare che i dati sintetici non siano semplici copie dei dati originali;
- permettere una piccola analisi tramite interfaccia utente.

Il progetto vuole quindi mostrare come l’intelligenza artificiale possa essere usata per creare dati artificiali utili per analisi, test o simulazioni, riducendo il rischio di esporre dati sensibili reali.

## 3. Dataset scelto

Per questo progetto verrà utilizzato un dataset pubblico tabellare.

Il dataset scelto sarà il **Pima Indians Diabetes Dataset**, che contiene dati medici relativi a pazienti e ad alcune caratteristiche utili per valutare il rischio di diabete.

Questo dataset è adatto al progetto perché:

- è tabellare;
- contiene dati numerici;
- è abbastanza semplice da gestire in Google Colab;
- rappresenta un caso vicino al tema dei dati sensibili;
- permette di confrontare facilmente dati reali e sintetici.

Anche se il dataset è pubblico, nel progetto verrà trattato come esempio di dato sensibile, perché in un caso reale informazioni mediche di questo tipo richiederebbero attenzione alla privacy.

## 4. Struttura generale del sistema

Il progetto sarà diviso in più parti.

La prima parte sarà dedicata alla preparazione dei dati. Il dataset verrà caricato, controllato e trasformato in un formato adatto al modello. In questa fase verranno gestiti eventuali valori mancanti e le variabili numeriche verranno normalizzate.

La seconda parte sarà dedicata alla costruzione del modello VAE. Il modello dovrà imparare una rappresentazione più compatta dei dati originali e poi usare questa rappresentazione per ricostruire e generare nuovi dati.

La terza parte sarà dedicata alla generazione del dataset sintetico. Dopo l’addestramento, il modello verrà usato per creare nuove righe di dati con caratteristiche simili a quelle del dataset originale.

La quarta parte sarà dedicata alla validazione. Verranno confrontati i dati reali e sintetici tramite statistiche, grafici e matrici di correlazione. Inoltre verrà fatto un controllo per verificare che non ci siano righe sintetiche identiche a quelle originali.

L’ultima parte sarà dedicata all’interfaccia. Verrà realizzata una semplice interfaccia per visualizzare i risultati e rendere il progetto più facile da usare.

## 5. Tecnologie utilizzate

Il progetto verrà sviluppato principalmente su **Google Colab**, utilizzando la GPU quando disponibile.

Le principali tecnologie previste sono:

- Python;
- Pandas per la gestione dei dati;
- NumPy per le operazioni numeriche;
- Scikit-learn per la preparazione dei dati;
- PyTorch per la costruzione e l’addestramento del VAE;
- Matplotlib e Seaborn per i grafici;
- Streamlit per una possibile interfaccia utente;
- FastAPI per una possibile struttura API del progetto.

L’integrazione con un LLM potrà essere simulata o sviluppata in forma controllata, in modo da permettere all’utente di fare semplici domande sui dati sintetici.

## 6. Fasi di sviluppo

### Fase 1: Preparazione dei dati

In questa fase verrà caricato il dataset scelto.  
I dati verranno analizzati per capire la struttura delle colonne, la presenza di valori mancanti e il tipo di variabili presenti.

Successivamente i dati verranno normalizzati, in modo da renderli più adatti all’addestramento del modello.

### Fase 2: Creazione del modello VAE

In questa fase verrà costruito il Variational Autoencoder.

Il modello sarà formato da due parti principali:

- un encoder, che comprime i dati originali in una rappresentazione più piccola;
- un decoder, che prova a ricostruire i dati partendo da questa rappresentazione.

Il modello verrà addestrato per imparare la struttura generale del dataset.

### Fase 3: Generazione dei dati sintetici

Dopo l’addestramento, il VAE verrà usato per generare nuovi dati.

Questi dati dovranno avere una forma simile al dataset originale, ma non dovranno essere una copia esatta delle righe reali.

Il risultato sarà un nuovo dataset sintetico esportabile in formato CSV.

### Fase 4: Confronto tra dati reali e sintetici

In questa fase verranno confrontati i dati originali con quelli generati.

Il confronto verrà fatto usando:

- statistiche descrittive;
- grafici delle distribuzioni;
- confronto delle correlazioni;
- controllo di eventuali righe duplicate o troppo simili.

Questa parte è importante perché permette di capire se il modello ha generato dati utili e realistici.

### Fase 5: Interfaccia utente

Nella fase finale verrà creata una semplice interfaccia per rendere il progetto più chiaro e utilizzabile.

L’interfaccia potrà permettere di:

- caricare un file CSV;
- avviare l’addestramento del modello;
- generare dati sintetici;
- visualizzare grafici e tabelle;
- fare semplici domande sui dati generati.

## 7. Controlli di privacy e sicurezza

Un aspetto importante del progetto è verificare che i dati sintetici non siano semplicemente una copia dei dati originali.

Per questo motivo verrà fatto un controllo per individuare eventuali righe identiche tra dataset reale e dataset sintetico.

Inoltre, nel progetto non verranno usati dati privati reali, ma solo dataset pubblici.  
Eventuali chiavi API, token o credenziali non verranno inseriti nel notebook o caricati su GitHub.
