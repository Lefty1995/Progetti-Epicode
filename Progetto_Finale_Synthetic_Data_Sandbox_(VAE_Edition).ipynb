# STEP 3 - Analisi iniziale e preparazione dei dati

In questo step analizziamo meglio il dataset e lo prepariamo per il modello VAE.

Alcune colonne mediche contengono valori pari a 0. In certi casi questi valori non rappresentano un dato realistico, ma possono indicare un valore mancante.  
Per questo motivo sostituiamo gli zeri nelle colonne mediche principali con valori mancanti e poi li riempiamo usando la mediana della colonna.

Successivamente normalizziamo tutte le colonne in un intervallo tra 0 e 1, così il dataset può essere usato correttamente dal modello VAE.

```python
# STEP 3 - Analisi iniziale e preparazione dei dati

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split

# Copia del dataset per la pulizia
df_clean = df.copy()

print("Dimensione iniziale del dataset:", df_clean.shape)

# Colonne dove il valore 0 può rappresentare un dato mancante/non realistico
columns_with_invalid_zero = [
    "Glucose",
    "BloodPressure",
    "SkinThickness",
    "Insulin",
    "BMI"
]

print("\nNumero di valori pari a 0 prima della pulizia:")
for col in columns_with_invalid_zero:
    print(f"{col}: {(df_clean[col] == 0).sum()}")

# Sostituzione degli zeri con NaN nelle colonne selezionate
df_clean[columns_with_invalid_zero] = df_clean[columns_with_invalid_zero].replace(0, np.nan)

print("\nValori mancanti dopo la sostituzione degli zeri:")
print(df_clean.isnull().sum())

# Riempimento dei valori mancanti con la mediana della colonna
for col in columns_with_invalid_zero:
    median_value = df_clean[col].median()
    df_clean[col] = df_clean[col].fillna(median_value)

print("\nValori mancanti dopo il riempimento con la mediana:")
print(df_clean.isnull().sum())

# Controllo statistiche dopo la pulizia
print("\nStatistiche descrittive del dataset pulito:")
display(df_clean.describe())

# Distribuzione della variabile Outcome
print("\nDistribuzione della colonna Outcome:")
print(df_clean["Outcome"].value_counts())

plt.figure(figsize=(5, 4))
df_clean["Outcome"].value_counts().sort_index().plot(kind="bar")
plt.title("Distribuzione della colonna Outcome")
plt.xlabel("Outcome")
plt.ylabel("Numero di record")
plt.xticks(rotation=0)
plt.show()

# Normalizzazione dei dati tra 0 e 1
scaler = MinMaxScaler()
data_scaled = scaler.fit_transform(df_clean)

# Creiamo un DataFrame normalizzato
df_scaled = pd.DataFrame(data_scaled, columns=df_clean.columns)

print("\nPrime 5 righe del dataset normalizzato:")
display(df_scaled.head())

print("\nValori minimi dopo la normalizzazione:")
print(df_scaled.min())

print("\nValori massimi dopo la normalizzazione:")
print(df_scaled.max())

# Preparazione dei dati per PyTorch
X = df_scaled.values.astype(np.float32)

# Divisione train/test
X_train, X_test = train_test_split(
    X,
    test_size=0.2,
    random_state=42
)

print("\nDataset pronto per il modello VAE.")
print("Dimensione X_train:", X_train.shape)
print("Dimensione X_test:", X_test.shape)

# Numero di colonne in input per il VAE
input_dim = X_train.shape[1]

print("Numero di variabili in input per il VAE:", input_dim)
```

# STEP 4 - Creazione del modello VAE

In questo step costruiamo il modello VAE, cioè il cuore generativo del progetto.

Il VAE è composto da due parti principali:

- Encoder: prende i dati originali e li comprime in una rappresentazione più piccola.
- Decoder: prende questa rappresentazione e prova a ricostruire i dati originali.


```python
# STEP 4 - Creazione del modello VAE

import torch
import torch.nn as nn
import torch.nn.functional as F

from torch.utils.data import TensorDataset, DataLoader

# Convertiamo i dati NumPy in tensori PyTorch
X_train_tensor = torch.tensor(X_train, dtype=torch.float32)
X_test_tensor = torch.tensor(X_test, dtype=torch.float32)

# Creiamo i DataLoader
batch_size = 64

train_dataset = TensorDataset(X_train_tensor)
test_dataset = TensorDataset(X_test_tensor)

train_loader = DataLoader(
    train_dataset,
    batch_size=batch_size,
    shuffle=True
)

test_loader = DataLoader(
    test_dataset,
    batch_size=batch_size,
    shuffle=False
)

print("DataLoader creati correttamente.")
print("Numero batch train:", len(train_loader))
print("Numero batch test:", len(test_loader))


# Definizione del modello VAE
class TabularVAE(nn.Module):
    def __init__(self, input_dim, hidden_dim=64, latent_dim=4):
        super(TabularVAE, self).__init__()

        # Encoder
        self.encoder_fc1 = nn.Linear(input_dim, hidden_dim)
        self.encoder_fc2 = nn.Linear(hidden_dim, 32)

        # Lo spazio latente viene descritto da media e log-varianza
        self.fc_mu = nn.Linear(32, latent_dim)
        self.fc_logvar = nn.Linear(32, latent_dim)

        # Decoder
        self.decoder_fc1 = nn.Linear(latent_dim, 32)
        self.decoder_fc2 = nn.Linear(32, hidden_dim)
        self.decoder_output = nn.Linear(hidden_dim, input_dim)

    def encode(self, x):
        x = F.relu(self.encoder_fc1(x))
        x = F.relu(self.encoder_fc2(x))

        mu = self.fc_mu(x)
        logvar = self.fc_logvar(x)

        return mu, logvar

    def reparameterize(self, mu, logvar):
        # Reparameterization Trick
        std = torch.exp(0.5 * logvar)
        epsilon = torch.randn_like(std)

        z = mu + epsilon * std

        return z

    def decode(self, z):
        z = F.relu(self.decoder_fc1(z))
        z = F.relu(self.decoder_fc2(z))

        # Sigmoid perché i dati sono normalizzati tra 0 e 1
        reconstructed = torch.sigmoid(self.decoder_output(z))

        return reconstructed

    def forward(self, x):
        mu, logvar = self.encode(x)
        z = self.reparameterize(mu, logvar)
        reconstructed = self.decode(z)

        return reconstructed, mu, logvar


# Funzione di loss del VAE
def vae_loss(reconstructed_x, original_x, mu, logvar, beta=0.01):
    # Errore di ricostruzione
    reconstruction_loss = F.mse_loss(
        reconstructed_x,
        original_x,
        reduction="sum"
    )

    # KL Divergence
    kl_divergence = -0.5 * torch.sum(
        1 + logvar - mu.pow(2) - logvar.exp()
    )

    # Loss totale
    total_loss = reconstruction_loss + beta * kl_divergence

    return total_loss, reconstruction_loss, kl_divergence


# Creazione del modello
latent_dim = 4

vae_model = TabularVAE(
    input_dim=input_dim,
    hidden_dim=64,
    latent_dim=latent_dim
).to(device)

print("\nModello VAE creato correttamente.")
print(vae_model)

# Test veloce su un batch per verificare che tutto funzioni
sample_batch = next(iter(train_loader))[0].to(device)

with torch.no_grad():
    reconstructed_batch, mu, logvar = vae_model(sample_batch)

print("\nTest del modello completato.")
print("Forma input batch:", sample_batch.shape)
print("Forma output ricostruito:", reconstructed_batch.shape)
print("Forma media spazio latente:", mu.shape)
print("Forma log-varianza spazio latente:", logvar.shape)
```

# STEP 5 - Addestramento del modello VAE

In questo step addestriamo il Variational Autoencoder sui dati normalizzati.

Durante l'addestramento il modello prova a ricostruire i dati originali.  
La loss è composta da due parti:

- Reconstruction Loss: misura quanto bene il modello riesce a ricostruire i dati.
- KL Divergence: aiuta il modello a creare uno spazio latente più regolare.

Alla fine dell'addestramento visualizziamo l'andamento della loss per capire se il modello sta imparando.

```python
# STEP 5 - Addestramento del modello VAE

import torch.optim as optim
import matplotlib.pyplot as plt

# Impostiamo un seed per rendere i risultati più riproducibili
torch.manual_seed(42)
np.random.seed(42)

# Iperparametri di addestramento
learning_rate = 0.001
num_epochs = 150
beta = 0.01

# Ottimizzatore
optimizer = optim.Adam(
    vae_model.parameters(),
    lr=learning_rate
)

# Liste per salvare l'andamento della loss
train_losses = []
train_reconstruction_losses = []
train_kl_losses = []

test_losses = []

print("Inizio addestramento del modello VAE...\n")

for epoch in range(1, num_epochs + 1):
    vae_model.train()

    total_train_loss = 0
    total_reconstruction_loss = 0
    total_kl_loss = 0

    for batch in train_loader:
        x_batch = batch[0].to(device)

        optimizer.zero_grad()

        reconstructed_batch, mu, logvar = vae_model(x_batch)

        loss, reconstruction_loss, kl_divergence = vae_loss(
            reconstructed_batch,
            x_batch,
            mu,
            logvar,
            beta=beta
        )

        loss.backward()
        optimizer.step()

        total_train_loss += loss.item()
        total_reconstruction_loss += reconstruction_loss.item()
        total_kl_loss += kl_divergence.item()

    # Loss media sul train set
    avg_train_loss = total_train_loss / len(train_loader.dataset)
    avg_reconstruction_loss = total_reconstruction_loss / len(train_loader.dataset)
    avg_kl_loss = total_kl_loss / len(train_loader.dataset)

    train_losses.append(avg_train_loss)
    train_reconstruction_losses.append(avg_reconstruction_loss)
    train_kl_losses.append(avg_kl_loss)

    # Valutazione sul test set
    vae_model.eval()
    total_test_loss = 0

    with torch.no_grad():
        for batch in test_loader:
            x_batch = batch[0].to(device)

            reconstructed_batch, mu, logvar = vae_model(x_batch)

            loss, reconstruction_loss, kl_divergence = vae_loss(
                reconstructed_batch,
                x_batch,
                mu,
                logvar,
                beta=beta
            )

            total_test_loss += loss.item()

    avg_test_loss = total_test_loss / len(test_loader.dataset)
    test_losses.append(avg_test_loss)

    # Stampa ogni 10 epoche
    if epoch == 1 or epoch % 10 == 0:
        print(
            f"Epoch [{epoch}/{num_epochs}] "
            f"Train Loss: {avg_train_loss:.4f} | "
            f"Test Loss: {avg_test_loss:.4f} | "
            f"Recon Loss: {avg_reconstruction_loss:.4f} | "
            f"KL: {avg_kl_loss:.4f}"
        )

print("\nAddestramento completato.")

# Grafico andamento loss
plt.figure(figsize=(8, 5))
plt.plot(train_losses, label="Train Loss")
plt.plot(test_losses, label="Test Loss")
plt.title("Andamento della Loss durante l'addestramento")
plt.xlabel("Epoche")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)
plt.show()

# Grafico separato Reconstruction Loss e KL Divergence
plt.figure(figsize=(8, 5))
plt.plot(train_reconstruction_losses, label="Reconstruction Loss")
plt.plot(train_kl_losses, label="KL Divergence")
plt.title("Componenti della Loss del VAE")
plt.xlabel("Epoche")
plt.ylabel("Valore medio")
plt.legend()
plt.grid(True)
plt.show()

# Salvataggio del modello addestrato
torch.save(vae_model.state_dict(), "vae_model.pth")

print("Modello VAE salvato come: vae_model.pth")
```

# STEP 6 - Generazione del dataset sintetico

In questo step usiamo il modello VAE addestrato per generare nuovi dati sintetici.

Il modello genera inizialmente dati normalizzati tra 0 e 1.  
Successivamente riportiamo questi dati nella scala originale usando lo stesso scaler usato nella fase di preparazione.

Alla fine creiamo un nuovo DataFrame sintetico e lo salviamo in formato CSV.

```python
# STEP 6 - Generazione del dataset sintetico

import torch
import numpy as np
import pandas as pd

# Mettiamo il modello in modalità valutazione
vae_model.eval()

# Numero di righe sintetiche da generare
num_synthetic_samples = len(df_clean)

print("Numero di record sintetici da generare:", num_synthetic_samples)

# Generazione di punti casuali nello spazio latente
with torch.no_grad():
    z = torch.randn(num_synthetic_samples, latent_dim).to(device)

    # Il decoder genera dati sintetici normalizzati
    synthetic_scaled = vae_model.decode(z).cpu().numpy()

# Sicurezza: limitiamo i valori tra 0 e 1
synthetic_scaled = np.clip(synthetic_scaled, 0, 1)

# Torniamo dalla scala normalizzata alla scala originale
synthetic_original_scale = scaler.inverse_transform(synthetic_scaled)

# Creiamo il DataFrame sintetico
df_synthetic = pd.DataFrame(
    synthetic_original_scale,
    columns=df_clean.columns
)

# Sistemiamo le colonne che nel dataset originale sono intere
integer_columns = [
    "Pregnancies",
    "Age",
    "Outcome"
]

for col in integer_columns:
    df_synthetic[col] = df_synthetic[col].round().astype(int)

# La colonna Outcome deve essere solo 0 o 1
df_synthetic["Outcome"] = df_synthetic["Outcome"].clip(0, 1)

# Sistemiamo eventuali valori negativi, per sicurezza
for col in df_synthetic.columns:
    df_synthetic[col] = df_synthetic[col].clip(lower=0)

# Salvataggio del dataset sintetico
df_synthetic.to_csv("synthetic_data.csv", index=False)

print("Dataset sintetico generato correttamente.")
print("Dimensione dataset sintetico:", df_synthetic.shape)

print("\nPrime 5 righe del dataset sintetico:")
display(df_synthetic.head())

print("\nStatistiche descrittive del dataset sintetico:")
display(df_synthetic.describe())

print("\nDistribuzione della colonna Outcome nel dataset sintetico:")
print(df_synthetic["Outcome"].value_counts())

print("\nFile salvato come: synthetic_data.csv")
```

# STEP 7 - Validazione dei dati sintetici

In questo step confrontiamo il dataset reale con quello sintetico.

L'obiettivo è capire se i dati generati dal VAE mantengono caratteristiche simili a quelle del dataset originale.

Per fare questo analizziamo:

- statistiche descrittive;
- confronto delle medie;
- distribuzioni delle variabili principali;
- matrici di correlazione;

# STEP 8 - Privacy Check

In questo step controlliamo che i dati sintetici non siano semplici copie dei dati reali.

Il controllo viene fatto in due modi:

- controllo di righe identiche tra dataset reale e dataset sintetico;
- controllo delle righe sintetiche troppo vicine alle righe reali.

Questo controllo non sostituisce una valutazione professionale completa sulla privacy, ma permette di fare una prima verifica utile per il progetto.

```python
# STEP 8 - Privacy Check

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.neighbors import NearestNeighbors

print("Inizio Privacy Check...")

# Dataset reale e sintetico
real_data = df_clean.copy()
synthetic_data = df_synthetic.copy()

# -------------------------------------------------
# 1. Controllo righe identiche
# -------------------------------------------------

# Arrotondiamo a 4 decimali per evitare problemi causati dai numeri float
real_rounded = real_data.round(4)
synthetic_rounded = synthetic_data.round(4)

# Creiamo una rappresentazione testuale di ogni riga
real_rows = set(real_rounded.astype(str).agg("|".join, axis=1))
synthetic_rows = set(synthetic_rounded.astype(str).agg("|".join, axis=1))

# Righe sintetiche identiche a righe reali
identical_rows = synthetic_rows.intersection(real_rows)
num_identical_rows = len(identical_rows)

print("\nControllo righe identiche:")
print("Numero di righe sintetiche identiche a righe reali:", num_identical_rows)

# -------------------------------------------------
# 2. Controllo duplicati interni nel dataset sintetico
# -------------------------------------------------

synthetic_duplicates = synthetic_rounded.duplicated().sum()

print("\nControllo duplicati nel dataset sintetico:")
print("Numero di righe duplicate nel dataset sintetico:", synthetic_duplicates)

# -------------------------------------------------
# 3. Controllo righe troppo simili tramite distanza
# -------------------------------------------------

# Usiamo lo stesso scaler già addestrato sui dati reali
real_scaled_check = scaler.transform(real_data)
synthetic_scaled_check = scaler.transform(synthetic_data)

# Per ogni riga sintetica cerchiamo la riga reale più vicina
nearest_neighbors = NearestNeighbors(
    n_neighbors=1,
    metric="euclidean"
)

nearest_neighbors.fit(real_scaled_check)

distances, indices = nearest_neighbors.kneighbors(synthetic_scaled_check)

# Convertiamo in array semplice
distances = distances.flatten()
indices = indices.flatten()

# Soglia scelta per individuare righe molto vicine
# Più la distanza è vicina a 0, più la riga sintetica è simile a una riga reale
similarity_threshold = 0.05

num_very_close_rows = np.sum(distances < similarity_threshold)

print("\nControllo righe sintetiche molto vicine a righe reali:")
print("Distanza minima trovata:", round(distances.min(), 6))
print("Distanza media:", round(distances.mean(), 6))
print("Distanza massima:", round(distances.max(), 6))
print(f"Numero di righe con distanza inferiore a {similarity_threshold}:", num_very_close_rows)

# -------------------------------------------------
# 4. Creazione report privacy
# -------------------------------------------------

privacy_report = pd.DataFrame({
    "metric": [
        "Righe sintetiche identiche a righe reali",
        "Righe duplicate nel dataset sintetico",
        "Distanza minima sintetico-reale",
        "Distanza media sintetico-reale",
        "Righe molto vicine sotto soglia"
    ],
    "value": [
        num_identical_rows,
        synthetic_duplicates,
        round(distances.min(), 6),
        round(distances.mean(), 6),
        int(num_very_close_rows)
    ]
})

print("\nReport Privacy Check:")
display(privacy_report)

privacy_report.to_csv("privacy_check_report.csv", index=False)

# -------------------------------------------------
# 5. Grafico delle distanze
# -------------------------------------------------

plt.figure(figsize=(8, 5))
plt.hist(distances, bins=30)
plt.axvline(similarity_threshold, linestyle="--", label=f"Soglia: {similarity_threshold}")
plt.title("Distribuzione delle distanze tra dati sintetici e dati reali")
plt.xlabel("Distanza dalla riga reale più vicina")
plt.ylabel("Numero di righe sintetiche")
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig("images/privacy_distance_distribution.png", dpi=300)
plt.show()

print("\nFile salvato:")
print("- privacy_check_report.csv")
print("- images/privacy_distance_distribution.png")

# -------------------------------------------------
# 6. Interpretazione semplice del risultato
# -------------------------------------------------

print("\nInterpretazione:")

if num_identical_rows == 0:
    print("- Non sono state trovate righe sintetiche identiche alle righe reali.")
else:
    print("- Attenzione: sono state trovate righe sintetiche identiche a righe reali.")

if num_very_close_rows == 0:
    print("- Non sono state trovate righe sintetiche troppo vicine secondo la soglia scelta.")
else:
    print("- Alcune righe sintetiche risultano molto vicine a righe reali. Questo va segnalato come limite del progetto.")

print("\nPrivacy Check completato.")
```

# STEP 9 - Natural Language Analyst

In questo step aggiungiamo una piccola funzione di analisi in linguaggio naturale.

L'utente può fare domande semplici sul dataset sintetico, per esempio:

- Qual è la correlazione tra Glucose e BMI?
- Qual è la media di Age?
- Mostrami la distribuzione di Outcome.
- Fammi un riassunto del dataset sintetico.

Per rendere il progetto più sicuro, il sistema non invia tutto il dataset all'LLM.  
Vengono usati solo i nomi delle colonne, alcune statistiche aggregate e il risultato calcolato localmente con Python.

```python
# STEP 9 - Natural Language Analyst

import os
import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Dataset usato dall'analyst
analyst_df = df_synthetic.copy()

print("Natural Language Analyst pronto.")
print("Dataset usato:", analyst_df.shape)
print("Colonne disponibili:")
print(analyst_df.columns.tolist())


# -------------------------------------------------
# Funzioni di supporto
# -------------------------------------------------

def find_columns_in_question(question, columns):
    """
    Cerca nel testo della domanda i nomi delle colonne presenti nel dataset.
    """
    question_lower = question.lower()
    found_columns = []

    for col in columns:
        if col.lower() in question_lower:
            found_columns.append(col)

    return found_columns


def local_dataset_analysis(question, data):
    """
    Esegue analisi semplici in locale usando Pandas.
    Questa parte funziona anche senza API OpenAI.
    """
    question_lower = question.lower()
    columns = data.columns.tolist()
    found_columns = find_columns_in_question(question, columns)

    result_text = ""
    chart_created = False

    # Caso 1: correlazione tra due colonne
    if "correlazione" in question_lower or "correlation" in question_lower:
        if len(found_columns) >= 2:
            col1 = found_columns[0]
            col2 = found_columns[1]

            corr_value = data[col1].corr(data[col2])

            result_text = (
                f"La correlazione tra {col1} e {col2} nel dataset sintetico "
                f"è pari a {corr_value:.4f}."
            )

            plt.figure(figsize=(6, 4))
            plt.scatter(data[col1], data[col2], alpha=0.6)
            plt.title(f"Correlazione tra {col1} e {col2}")
            plt.xlabel(col1)
            plt.ylabel(col2)
            plt.grid(alpha=0.3)
            plt.tight_layout()
            plt.show()

            chart_created = True

        else:
            result_text = (
                "Hai chiesto una correlazione, ma non ho trovato due colonne valide nella domanda. "
                "Esempio: Qual è la correlazione tra Glucose e BMI?"
            )

    # Caso 2: media di una colonna
    elif "media" in question_lower or "mean" in question_lower or "valore medio" in question_lower:
        if len(found_columns) >= 1:
            col = found_columns[0]
            mean_value = data[col].mean()

            result_text = (
                f"La media della colonna {col} nel dataset sintetico "
                f"è pari a {mean_value:.4f}."
            )
        else:
            result_text = (
                "Hai chiesto una media, ma non ho trovato una colonna valida nella domanda. "
                "Esempio: Qual è la media di Age?"
            )

    # Caso 3: minimo
    elif "minimo" in question_lower or "minimum" in question_lower:
        if len(found_columns) >= 1:
            col = found_columns[0]
            min_value = data[col].min()

            result_text = (
                f"Il valore minimo della colonna {col} nel dataset sintetico "
                f"è pari a {min_value:.4f}."
            )
        else:
            result_text = "Non ho trovato una colonna valida per calcolare il minimo."

    # Caso 4: massimo
    elif "massimo" in question_lower or "maximum" in question_lower:
        if len(found_columns) >= 1:
            col = found_columns[0]
            max_value = data[col].max()

            result_text = (
                f"Il valore massimo della colonna {col} nel dataset sintetico "
                f"è pari a {max_value:.4f}."
            )
        else:
            result_text = "Non ho trovato una colonna valida per calcolare il massimo."

    # Caso 5: distribuzione
    elif "distribuzione" in question_lower or "distribution" in question_lower:
        if len(found_columns) >= 1:
            col = found_columns[0]

            result_text = f"Ho generato la distribuzione della colonna {col}."

            plt.figure(figsize=(7, 4))

            if data[col].nunique() <= 10:
                data[col].value_counts().sort_index().plot(kind="bar")
                plt.xlabel(col)
                plt.ylabel("Numero di record")
            else:
                plt.hist(data[col], bins=30)
                plt.xlabel(col)
                plt.ylabel("Frequenza")

            plt.title(f"Distribuzione di {col}")
            plt.grid(alpha=0.3)
            plt.tight_layout()
            plt.show()

            chart_created = True

        else:
            result_text = (
                "Hai chiesto una distribuzione, ma non ho trovato una colonna valida nella domanda. "
                "Esempio: Mostrami la distribuzione di Outcome."
            )

    # Caso 6: riassunto generale
    elif "riassunto" in question_lower or "summary" in question_lower or "descrivi" in question_lower:
        result_text = (
            f"Il dataset sintetico contiene {data.shape[0]} righe e {data.shape[1]} colonne. "
            f"Le colonne sono: {', '.join(columns)}. "
            "Il dataset rappresenta una versione sintetica dei dati medici originali generata tramite VAE."
        )

    # Caso fallback
    else:
        result_text = (
            "Non ho riconosciuto completamente la domanda. "
            "Puoi provare con domande come: "
            "'Qual è la correlazione tra Glucose e BMI?', "
            "'Qual è la media di Age?' oppure "
            "'Mostrami la distribuzione di Outcome'."
        )

    return result_text, chart_created


def create_dataset_context(data):
    """
    Crea un contesto sintetico da dare all'LLM.
    Non invia tutto il dataset, ma solo informazioni aggregate.
    """
    summary_stats = data.describe().round(3).to_string()
    correlation_matrix = data.corr().round(3).to_string()

    context = f"""
Dataset sintetico generato con VAE.

Numero righe: {data.shape[0]}
Numero colonne: {data.shape[1]}

Colonne:
{', '.join(data.columns.tolist())}

Statistiche descrittive:
{summary_stats}

Matrice di correlazione:
{correlation_matrix}
"""
    return context


def improve_answer_with_llm(question, local_answer, data):
    """
    Usa OpenAI solo per rendere la risposta più chiara.
    Se la chiave API non funziona, restituisce la risposta locale.
    """
    api_key = os.environ.get("OPENAI_API_KEY")

    if not api_key:
        return local_answer + "\n\nNota: risposta generata senza LLM perché la chiave API non è disponibile."

    try:
        from openai import OpenAI

        client = OpenAI(api_key=api_key)

        dataset_context = create_dataset_context(data)

        response = client.responses.create(
            model="gpt-4.1-mini",
            instructions=(
                "Sei un assistente di analisi dati. "
                "Rispondi in italiano in modo semplice, chiaro e adatto a uno studente. "
                "Usa solo le informazioni fornite nel contesto. "
                "Non inventare dati. "
                "Non dire di aver visto il dataset completo, perché ricevi solo statistiche aggregate."
            ),
            input=f"""
Domanda dell'utente:
{question}

Risultato calcolato localmente con Python:
{local_answer}

Contesto aggregato del dataset:
{dataset_context}

Scrivi una risposta breve e comprensibile.
"""
        )

        return response.output_text

    except Exception as e:
        return (
            local_answer
            + "\n\nNota: la risposta LLM non è stata generata. "
            + f"Motivo tecnico: {str(e)}"
        )


def ask_dataset_analyst(question, use_llm=True):
    """
    Funzione principale da usare per fare domande sul dataset sintetico.
    """
    print("Domanda:")
    print(question)

    print("\nAnalisi locale:")
    local_answer, chart_created = local_dataset_analysis(question, analyst_df)
    print(local_answer)

    if use_llm:
        print("\nRisposta migliorata con LLM:")
        final_answer = improve_answer_with_llm(question, local_answer, analyst_df)
        print(final_answer)

    return local_answer


# -------------------------------------------------
# Test dello Step 9
# -------------------------------------------------

print("\nEsempio di domanda al Natural Language Analyst:\n")

ask_dataset_analyst(
    "Qual è la correlazione tra Glucose e BMI?",
    use_llm=True
)
```

```python
# STEP 10 - Creazione dell'interfaccia Streamlit

# Salviamo anche il dataset reale pulito, così Streamlit può leggerlo
df_clean.to_csv("real_clean_data.csv", index=False)

print("File real_clean_data.csv salvato correttamente.")

# Creazione del file app.py
app_code = r'''
import os
import pandas as pd
import numpy as np
import streamlit as st
import matplotlib.pyplot as plt

st.set_page_config(
    page_title="Synthetic Data Sandbox - VAE Edition",
    page_icon="🧬",
    layout="wide"
)

# -------------------------------------------------
# Funzioni di caricamento
# -------------------------------------------------

@st.cache_data
def load_csv(path):
    if os.path.exists(path):
        return pd.read_csv(path)
    return None

real_data = load_csv("real_clean_data.csv")
synthetic_data = load_csv("synthetic_data.csv")
statistics_comparison = load_csv("statistics_comparison.csv")
privacy_report = load_csv("privacy_check_report.csv")

# -------------------------------------------------
# Titolo principale
# -------------------------------------------------

st.title("Synthetic Data Sandbox - VAE Edition")
st.write(
    "Dashboard dimostrativa per generare, confrontare e analizzare dati sintetici "
    "creati con un modello VAE."
)

st.info(
    "Il progetto usa un dataset medico pubblico come esempio di dati sensibili. "
    "Il dataset sintetico viene generato per mantenere proprietà simili ai dati originali, "
    "senza copiare direttamente le righe reali."
)

# -------------------------------------------------
# Sidebar
# -------------------------------------------------

st.sidebar.title("Menu")
section = st.sidebar.radio(
    "Scegli una sezione",
    [
        "Overview",
        "Dataset",
        "Validazione",
        "Privacy Check",
        "Natural Language Analyst",
        "Upload CSV Demo"
    ]
)

# -------------------------------------------------
# Overview
# -------------------------------------------------

if section == "Overview":
    st.header("Overview del progetto")

    col1, col2, col3 = st.columns(3)

    if real_data is not None:
        col1.metric("Righe dataset reale", real_data.shape[0])
        col2.metric("Colonne dataset reale", real_data.shape[1])

    if synthetic_data is not None:
        col3.metric("Righe dataset sintetico", synthetic_data.shape[0])

    st.subheader("Obiettivo")
    st.write(
        "L'obiettivo del progetto è creare una piattaforma che parta da un dataset tabellare, "
        "addestri un Variational Autoencoder e generi un nuovo dataset sintetico. "
        "Successivamente i dati sintetici vengono confrontati con quelli reali tramite statistiche, "
        "grafici e controlli di privacy."
    )

    st.subheader("Componenti principali")
    st.markdown(
        """
        - **Data Engine:** preparazione e normalizzazione del dataset.
        - **Generative Core:** modello VAE per generare dati sintetici.
        - **Validation:** confronto tra dati reali e sintetici.
        - **Privacy Check:** controllo di righe identiche o troppo simili.
        - **User Interface:** dashboard Streamlit per visualizzare i risultati.
        - **Natural Language Analyst:** piccola funzione per interrogare i dati sintetici.
        """
    )

# -------------------------------------------------
# Dataset
# -------------------------------------------------

elif section == "Dataset":
    st.header("Dataset reale e sintetico")

    if real_data is None or synthetic_data is None:
        st.error("File del dataset non trovati. Assicurati di aver eseguito gli step precedenti.")
    else:
        tab1, tab2, tab3 = st.tabs(
            ["Dataset reale pulito", "Dataset sintetico", "Statistiche"]
        )

        with tab1:
            st.subheader("Dataset reale pulito")
            st.dataframe(real_data.head(50), use_container_width=True)
            st.write("Dimensione:", real_data.shape)

        with tab2:
            st.subheader("Dataset sintetico")
            st.dataframe(synthetic_data.head(50), use_container_width=True)
            st.write("Dimensione:", synthetic_data.shape)

            csv = synthetic_data.to_csv(index=False).encode("utf-8")
            st.download_button(
                label="Scarica dataset sintetico CSV",
                data=csv,
                file_name="synthetic_data.csv",
                mime="text/csv"
            )

        with tab3:
            st.subheader("Statistiche dataset reale")
            st.dataframe(real_data.describe(), use_container_width=True)

            st.subheader("Statistiche dataset sintetico")
            st.dataframe(synthetic_data.describe(), use_container_width=True)

# -------------------------------------------------
# Validazione
# -------------------------------------------------

elif section == "Validazione":
    st.header("Validazione dati reali vs sintetici")

    if statistics_comparison is not None:
        st.subheader("Confronto statistiche principali")
        st.dataframe(statistics_comparison, use_container_width=True)
    else:
        st.warning("File statistics_comparison.csv non trovato.")

    st.subheader("Grafici di validazione")

    image_paths = [
        "images/outcome_distribution_comparison.png",
        "images/distribution_comparison_Glucose.png",
        "images/distribution_comparison_BMI.png",
        "images/distribution_comparison_Age.png",
        "images/correlation_real.png",
        "images/correlation_synthetic.png",
        "images/correlation_difference.png"
    ]

    for path in image_paths:
        if os.path.exists(path):
            st.image(path, caption=path)
        else:
            st.warning(f"Immagine non trovata: {path}")

# -------------------------------------------------
# Privacy Check
# -------------------------------------------------

elif section == "Privacy Check":
    st.header("Privacy Check")

    st.write(
        "Questa sezione controlla se il dataset sintetico contiene righe identiche "
        "o troppo vicine alle righe reali."
    )

    if privacy_report is not None:
        st.subheader("Report privacy")
        st.dataframe(privacy_report, use_container_width=True)
    else:
        st.warning("File privacy_check_report.csv non trovato.")

    privacy_image = "images/privacy_distance_distribution.png"

    if os.path.exists(privacy_image):
        st.subheader("Distribuzione delle distanze")
        st.image(privacy_image)
    else:
        st.warning("Grafico privacy non trovato.")

    st.info(
        "Nel progetto finale questo controllo viene usato come verifica base. "
        "Non sostituisce una valutazione completa di privacy, ma aiuta a capire "
        "se il modello sta copiando direttamente i dati reali."
    )

# -------------------------------------------------
# Natural Language Analyst
# -------------------------------------------------

elif section == "Natural Language Analyst":
    st.header("Natural Language Analyst")

    if synthetic_data is None:
        st.error("Dataset sintetico non trovato.")
    else:
        st.write(
            "Fai una domanda semplice sul dataset sintetico. "
            "Esempi: correlazione, media, minimo, massimo o distribuzione."
        )

        st.markdown(
            """
            **Esempi di domande:**

            - Qual è la correlazione tra Glucose e BMI?
            - Qual è la media di Age?
            - Mostrami la distribuzione di Outcome.
            - Qual è il massimo di Insulin?
            - Fammi un riassunto del dataset.
            """
        )

        question = st.text_input(
            "Scrivi la tua domanda",
            value="Qual è la correlazione tra Glucose e BMI?"
        )

        use_llm = st.checkbox(
            "Usa LLM per migliorare la risposta se la chiave API è disponibile",
            value=False
        )

        def find_columns_in_question(question, columns):
            question_lower = question.lower()
            found_columns = []

            for col in columns:
                if col.lower() in question_lower:
                    found_columns.append(col)

            return found_columns

        def local_analysis(question, data):
            question_lower = question.lower()
            columns = data.columns.tolist()
            found_columns = find_columns_in_question(question, columns)

            result_text = ""
            fig = None

            if "correlazione" in question_lower or "correlation" in question_lower:
                if len(found_columns) >= 2:
                    col1 = found_columns[0]
                    col2 = found_columns[1]
                    corr_value = data[col1].corr(data[col2])

                    result_text = (
                        f"La correlazione tra {col1} e {col2} nel dataset sintetico "
                        f"è pari a {corr_value:.4f}."
                    )

                    fig, ax = plt.subplots(figsize=(6, 4))
                    ax.scatter(data[col1], data[col2], alpha=0.6)
                    ax.set_title(f"Correlazione tra {col1} e {col2}")
                    ax.set_xlabel(col1)
                    ax.set_ylabel(col2)
                    ax.grid(alpha=0.3)

                else:
                    result_text = (
                        "Hai chiesto una correlazione, ma non ho trovato due colonne valide. "
                        "Esempio: Qual è la correlazione tra Glucose e BMI?"
                    )

            elif "media" in question_lower or "mean" in question_lower or "valore medio" in question_lower:
                if len(found_columns) >= 1:
                    col = found_columns[0]
                    mean_value = data[col].mean()
                    result_text = f"La media della colonna {col} è pari a {mean_value:.4f}."
                else:
                    result_text = "Non ho trovato una colonna valida per calcolare la media."

            elif "minimo" in question_lower or "minimum" in question_lower:
                if len(found_columns) >= 1:
                    col = found_columns[0]
                    min_value = data[col].min()
                    result_text = f"Il valore minimo della colonna {col} è pari a {min_value:.4f}."
                else:
                    result_text = "Non ho trovato una colonna valida per calcolare il minimo."

            elif "massimo" in question_lower or "maximum" in question_lower:
                if len(found_columns) >= 1:
                    col = found_columns[0]
                    max_value = data[col].max()
                    result_text = f"Il valore massimo della colonna {col} è pari a {max_value:.4f}."
                else:
                    result_text = "Non ho trovato una colonna valida per calcolare il massimo."

            elif "distribuzione" in question_lower or "distribution" in question_lower:
                if len(found_columns) >= 1:
                    col = found_columns[0]
                    result_text = f"Distribuzione della colonna {col}."

                    fig, ax = plt.subplots(figsize=(6, 4))

                    if data[col].nunique() <= 10:
                        data[col].value_counts().sort_index().plot(kind="bar", ax=ax)
                        ax.set_ylabel("Numero di record")
                    else:
                        ax.hist(data[col], bins=30)
                        ax.set_ylabel("Frequenza")

                    ax.set_title(f"Distribuzione di {col}")
                    ax.set_xlabel(col)
                    ax.grid(alpha=0.3)

                else:
                    result_text = "Non ho trovato una colonna valida per mostrare la distribuzione."

            elif "riassunto" in question_lower or "summary" in question_lower or "descrivi" in question_lower:
                result_text = (
                    f"Il dataset sintetico contiene {data.shape[0]} righe e {data.shape[1]} colonne. "
                    f"Le colonne sono: {', '.join(columns)}. "
                    "È una versione sintetica del dataset medico originale generata tramite VAE."
                )

            else:
                result_text = (
                    "Non ho riconosciuto completamente la domanda. "
                    "Prova con una domanda su correlazione, media, minimo, massimo o distribuzione."
                )

            return result_text, fig

        def improve_with_llm(question, local_answer, data):
            api_key = os.environ.get("OPENAI_API_KEY")

            if not api_key:
                return local_answer + "\n\nNota: chiave API non disponibile nell'ambiente Streamlit."

            try:
                from openai import OpenAI

                client = OpenAI(api_key=api_key)

                summary_stats = data.describe().round(3).to_string()
                corr_matrix = data.corr().round(3).to_string()

                response = client.responses.create(
                    model="gpt-4.1-mini",
                    instructions=(
                        "Sei un assistente di analisi dati. "
                        "Rispondi in italiano, in modo semplice e chiaro. "
                        "Non inventare dati. Usa solo il risultato locale e il contesto aggregato."
                    ),
                    input=f"""
Domanda:
{question}

Risultato calcolato localmente:
{local_answer}

Statistiche aggregate:
{summary_stats}

Correlazioni aggregate:
{corr_matrix}

Rispondi in modo breve e comprensibile.
"""
                )

                return response.output_text

            except Exception as e:
                return local_answer + f"\n\nNota: LLM non disponibile. Errore: {str(e)}"

        if st.button("Analizza domanda"):
            local_answer, fig = local_analysis(question, synthetic_data)

            st.subheader("Risposta")
            st.write(local_answer)

            if fig is not None:
                st.pyplot(fig)

            if use_llm:
                st.subheader("Risposta migliorata con LLM")
                final_answer = improve_with_llm(question, local_answer, synthetic_data)
                st.write(final_answer)

# -------------------------------------------------
# Upload CSV Demo
# -------------------------------------------------

elif section == "Upload CSV Demo":
    st.header("Upload CSV Demo")

    st.write(
        "Questa sezione mostra la parte di caricamento file richiesta dalla piattaforma. "
        "L'utente può caricare un CSV e visualizzarne una prima analisi."
    )

    uploaded_file = st.file_uploader("Carica un file CSV", type=["csv"])

    if uploaded_file is not None:
        uploaded_df = pd.read_csv(uploaded_file)

        st.success("File CSV caricato correttamente.")
        st.write("Dimensione del file:", uploaded_df.shape)

        st.subheader("Prime righe")
        st.dataframe(uploaded_df.head(), use_container_width=True)

        st.subheader("Statistiche descrittive")
        st.dataframe(uploaded_df.describe(), use_container_width=True)
    else:
        st.info("Carica un file CSV per visualizzarlo nella dashboard.")
'''

with open("app.py", "w", encoding="utf-8") as f:
    f.write(app_code)

print("File app.py creato correttamente.")
print("Questo file contiene la dashboard Streamlit del progetto.")
```

# STEP 12 - Creazione del Backend FastAPI

In questo step creiamo un semplice backend con FastAPI.

Il backend espone alcuni endpoint principali:

- `/` per controllare se l'API è attiva;
- `/upload` per caricare un file CSV;
- `/train` per simulare l'avvio dell'addestramento del modello VAE;
- `/generate` per scaricare il dataset sintetico generato;
- `/privacy-report` per leggere il report del controllo privacy;
- `/statistics` per leggere il confronto statistico tra dati reali e sintetici.

Questa parte rappresenta l'Analytic Backend del progetto.

```python
# STEP 12 - Creazione del Backend FastAPI

api_code = r'''
import os
import pandas as pd

from fastapi import FastAPI, UploadFile, File
from fastapi.responses import FileResponse, JSONResponse

app = FastAPI(
    title="Synthetic Data Sandbox API",
    description="Backend API per il progetto Synthetic Data Sandbox - VAE Edition",
    version="1.0"
)

# -------------------------------------------------
# Endpoint base
# -------------------------------------------------

@app.get("/")
def home():
    return {
        "message": "Synthetic Data Sandbox API attiva",
        "project": "VAE Edition",
        "available_endpoints": [
            "/upload",
            "/train",
            "/generate",
            "/privacy-report",
            "/statistics"
        ]
    }


# -------------------------------------------------
# Upload CSV
# -------------------------------------------------

@app.post("/upload")
async def upload_csv(file: UploadFile = File(...)):
    """
    Endpoint per caricare un file CSV.
    Il file viene salvato come uploaded_dataset.csv.
    """

    if not file.filename.endswith(".csv"):
        return JSONResponse(
            status_code=400,
            content={"error": "Il file deve essere in formato CSV."}
        )

    file_path = "uploaded_dataset.csv"

    with open(file_path, "wb") as f:
        content = await file.read()
        f.write(content)

    try:
        df = pd.read_csv(file_path)

        return {
            "message": "File CSV caricato correttamente.",
            "filename": file.filename,
            "saved_as": file_path,
            "rows": df.shape[0],
            "columns": df.shape[1],
            "column_names": df.columns.tolist()
        }

    except Exception as e:
        return JSONResponse(
            status_code=500,
            content={"error": f"Errore durante la lettura del CSV: {str(e)}"}
        )


# -------------------------------------------------
# Train
# -------------------------------------------------

@app.post("/train")
def train_model():
    """
    Endpoint dimostrativo per rappresentare l'avvio dell'addestramento del VAE.

    Nel notebook Colab l'addestramento vero è già stato eseguito nello Step 5.
    Questo endpoint serve a mostrare come il backend gestirebbe la richiesta.
    """

    model_path = "vae_model.pth"

    if os.path.exists(model_path):
        return {
            "message": "Modello VAE già addestrato e disponibile.",
            "model_file": model_path,
            "status": "ready"
        }

    return {
        "message": "Il modello non è ancora stato trovato.",
        "status": "missing_model",
        "suggestion": "Esegui prima lo Step 5 nel notebook Colab."
    }


# -------------------------------------------------
# Generate
# -------------------------------------------------

@app.get("/generate")
def generate_synthetic_data():
    """
    Endpoint per scaricare il dataset sintetico generato.
    """

    synthetic_path = "synthetic_data.csv"

    if not os.path.exists(synthetic_path):
        return JSONResponse(
            status_code=404,
            content={
                "error": "Dataset sintetico non trovato.",
                "suggestion": "Esegui prima lo Step 6 nel notebook Colab."
            }
        )

    return FileResponse(
        path=synthetic_path,
        filename="synthetic_data.csv",
        media_type="text/csv"
    )


# -------------------------------------------------
# Privacy Report
# -------------------------------------------------

@app.get("/privacy-report")
def get_privacy_report():
    """
    Endpoint per visualizzare il report del Privacy Check.
    """

    report_path = "privacy_check_report.csv"

    if not os.path.exists(report_path):
        return JSONResponse(
            status_code=404,
            content={
                "error": "Privacy report non trovato.",
                "suggestion": "Esegui prima lo Step 8 nel notebook Colab."
            }
        )

    report = pd.read_csv(report_path)

    return {
        "message": "Privacy report caricato correttamente.",
        "report": report.to_dict(orient="records")
    }


# -------------------------------------------------
# Statistics
# -------------------------------------------------

@app.get("/statistics")
def get_statistics_comparison():
    """
    Endpoint per visualizzare il confronto statistico tra dati reali e sintetici.
    """

    stats_path = "statistics_comparison.csv"

    if not os.path.exists(stats_path):
        return JSONResponse(
            status_code=404,
            content={
                "error": "File statistics_comparison.csv non trovato.",
                "suggestion": "Esegui prima lo Step 7 nel notebook Colab."
            }
        )

    stats = pd.read_csv(stats_path)

    return {
        "message": "Statistiche caricate correttamente.",
        "statistics": stats.to_dict(orient="records")
    }


# -------------------------------------------------
# Dataset Info
# -------------------------------------------------

@app.get("/dataset-info")
def get_dataset_info():
    """
    Endpoint per leggere informazioni base sui dataset del progetto.
    """

    real_path = "real_clean_data.csv"
    synthetic_path = "synthetic_data.csv"

    result = {}

    if os.path.exists(real_path):
        real_df = pd.read_csv(real_path)
        result["real_dataset"] = {
            "rows": real_df.shape[0],
            "columns": real_df.shape[1],
            "column_names": real_df.columns.tolist()
        }
    else:
        result["real_dataset"] = "File real_clean_data.csv non trovato."

    if os.path.exists(synthetic_path):
        synthetic_df = pd.read_csv(synthetic_path)
        result["synthetic_dataset"] = {
            "rows": synthetic_df.shape[0],
            "columns": synthetic_df.shape[1],
            "column_names": synthetic_df.columns.tolist()
        }
    else:
        result["synthetic_dataset"] = "File synthetic_data.csv non trovato."

    return result
'''

with open("api.py", "w", encoding="utf-8") as f:
    f.write(api_code)

print("File api.py creato correttamente.")
print("Backend FastAPI pronto.")
```

```python
import os
import shutil

# Cartella dove sono i file
source_folder = "/content"

# Cartella finale pulita
final_folder = "/content/vae_synthetic_project_final_clean"

# Elimina la cartella finale se esiste già
if os.path.exists(final_folder):
    shutil.rmtree(final_folder)

os.makedirs(final_folder, exist_ok=True)

# File da copiare nel progetto finale
files_to_copy = [
    "README.md",
    "PROJECT_SUMMARY.md",
    "requirements.txt",
    ".gitignore",
    "app.py",
    "api.py",
    "vae_model.pth",
    "real_clean_data.csv",
    "synthetic_data.csv",
    "statistics_comparison.csv",
    "privacy_check_report.csv"
]

# Copia i file principali
for file_name in files_to_copy:
    src = os.path.join(source_folder, file_name)
    dst = os.path.join(final_folder, file_name)

    if os.path.exists(src):
        shutil.copy(src, dst)
        print(f"OK copiato: {file_name}")
    else:
        print(f"ATTENZIONE manca: {file_name}")

# Copia la cartella images
images_src = os.path.join(source_folder, "images")
images_dst = os.path.join(final_folder, "images")

if os.path.exists(images_src):
    shutil.copytree(images_src, images_dst)
    print("OK copiata cartella images")
else:
    print("ATTENZIONE manca cartella images")

# Crea ZIP finale
zip_path = "/content/vae_synthetic_project_final_clean"
shutil.make_archive(zip_path, "zip", final_folder)

print("\nZIP creato correttamente:")
print(zip_path + ".zip")
```
