# REAL ESTATE AI ASSISTANT

## Sinossi del progetto

**REAL ESTATE AI ASSISTANT** è una piattaforma basata su Intelligenza Artificiale progettata per supportare le agenzie immobiliari nella trasformazione automatica di un annuncio immobiliare in materiali commerciali e contenuti marketing brandizzati.

Il sistema parte dal link di un annuncio online, oppure dal testo e dalle immagini dell'immobile, ed è in grado di estrarre automaticamente le informazioni principali, analizzare le fotografie e generare diversi output pronti all'uso, come:

- brochure PDF
- post social
- caroselli
- reel semplici

L'obiettivo del progetto è automatizzare un processo oggi spesso manuale, ripetitivo e dispendioso in termini di tempo: la rielaborazione di un annuncio pubblicato sui portali immobiliari in contenuti piu efficaci, professionali e adatti ai diversi canali di comunicazione.

In molte agenzie immobiliari, infatti, il testo dell'annuncio e le immagini vengono riutilizzati manualmente per creare:

- brochure stampabili da consegnare ai clienti
- post per Facebook o Instagram
- materiali promozionali coerenti con l'identita visiva dell'agenzia

Questo flusso di lavoro richiede tempo, competenze trasversali e una continua attivita di adattamento dei contenuti.

## Problema che il progetto risolve

Il problema che il progetto intende risolvere e la mancanza di uno strumento intelligente in grado di semplificare e automatizzare la creazione di materiali commerciali immobiliari a partire da una singola fonte, mantenendo:

- coerenza visiva
- qualita del contenuto
- adattamento ai diversi formati

Il sistema vuole:

- ridurre il tempo operativo dell'agente
- migliorare la qualita della presentazione degli immobili
- rendere piu efficiente la produzione di contenuti destinati sia alla stampa sia ai social media

## Funzionalita principali

Dal punto di vista funzionale, **REAL ESTATE AI ASSISTANT** prevede una fase iniziale di onboarding in cui l'utente:

- carica il logo dell'agenzia
- inserisce il nome del brand

A partire dal logo, il sistema estrae automaticamente i colori principali, che vengono poi utilizzati come palette visiva per tutti gli output, con possibilita di modifica manuale.

Successivamente l'utente puo:

- fornire il link dell'annuncio
- caricare testo e immagini manualmente

Il sistema analizza il contenuto dell'annuncio ed estrae dati strutturati come:

- tipologia dell'immobile
- localita
- prezzo
- metratura
- numero di locali
- bagni
- caratteristiche distintive
- punti di forza

## Analisi delle immagini

Una seconda componente del progetto riguarda l'analisi delle immagini attraverso tecniche di **Computer Vision**.

Le fotografie dell'immobile vengono classificate in categorie come:

- cucina
- bagno
- soggiorno
- camera
- esterno
- facciata

Oltre alla classificazione dell'ambiente, il sistema valuta anche la qualita visiva delle immagini per selezionare automaticamente quelle piu efficaci dal punto di vista commerciale.

Le immagini vengono poi ordinate secondo una logica narrativa utile alla presentazione dell'immobile.

## Output generati

A partire da queste informazioni, l'utente puo scegliere quali contenuti generare. Il sistema include la creazione automatica di:

- brochure PDF brandizzate
- post social per Facebook, Instagram e LinkedIn con copy differenziati per piattaforma
- post singoli con immagine scelta automaticamente
- caroselli con testi brevi associati alle foto degli ambienti
- un modulo chiamato **AI Reel Composer**, che realizza reel semplici in formato verticale utilizzando le immagini migliori, testi overlay e branding dell'agenzia

## Tecnologie e competenze coinvolte

Il progetto integra diverse aree studiate durante il corso:

- **NLP** per l'estrazione e la rielaborazione delle informazioni contenute nell'annuncio
- **Generative AI** per la produzione di testi adattati ai diversi output
- **Computer Vision** per il riconoscimento degli ambienti e la selezione automatica delle immagini

In questo modo il progetto si colloca all'intersezione tra:

- automazione documentale
- marketing immobiliare
- intelligenza artificiale multimodale

## Tecnologie previste

Le tecnologie previste per la realizzazione includono:

- **Python** come linguaggio principale
- **Google Colab** come ambiente di sviluppo
- librerie come `Pandas`, `OpenCV`, `Pillow`, `transformers`, `scikit-learn`
